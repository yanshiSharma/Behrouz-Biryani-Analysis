```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
b_url = "https://www.behrouzbiryani.com/chennai/vandalur/shahi-biryanis"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
b_r = requests.get(b_url, headers=headers)

print(b_r.status_code)
print(b_r.text)
```

```python
soup = BeautifulSoup(b_r.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
body = soup.find('body')
```

```python
print(body.prettify())
```

```python
root = body.find('div', id="react-root")
print(root.prettify())
```

```python
pattern1 = re.compile(r'^stickyHeader')
pattern2 = re.compile(r'^style__BrandCarouselHorizontalWrapper')
pattern3 = re.compile(r'^style__BrandCarouselContainer')
pattern4 = re.compile(r'^style__Slider')
pattern5 = re.compile(r'^style__BrandCard')
pattern6 = re.compile(r'^style__CardText')

p1 = root.find('div', class_=pattern1)
p2 = p1.find('div', class_=pattern2)
p3 = p2.find('div', class_=pattern3)
p4 = p3.find('div', class_=pattern4)

a = p4.find_all('a')
categories = []
for a_s in a:
    cat = a_s.find('div', class_=pattern5)
    category = a_s.find('div', class_=pattern6).text
    categories.append(category)
    
```

```python
pattern7 = re.compile(r'^style__Wrapper')
pattern8 = re.compile(r'^style__Container')
pattern9 = re.compile(r'^sc-hKwCoD')

p5 = p1.find('div', class_=pattern7)
p6 = p5.find('div', class_=pattern8)
p7 = p6.find('div', class_=pattern9)
###p8 = p7.find('div', class_=pattern8)

p8 = p7.find_all('div')
for p in p8:
    p9 = p.find('div', class_="^style__CategorySection")
    if not p9:
        print(p.prettify())
        print("\n\n\n\n\n\n\n\n\n")
    else:
        pass
#print(p7.prettify())

```

```python
#####################################################SHAHI-BIRYANI##############################################################
for p in p8:
    # Find all product containers
    product_containers = soup.find_all('figure', class_="style__Card-sc-1xxctwg-0 exocxh")

    # Extract information from each product
    products = []
    for container in product_containers:
        # Get product name
        name_div = container.find('div', class_="style__Title-sc-1xxctwg-8 dmwYVP")
        name = name_div.text.strip() if name_div else "N/A"

        # Get product description
        desc_p = container.find('p', class_="style__Desc-sc-1xxctwg-9 hhfuSv")
        description = desc_p.text.strip() if desc_p else "N/A"

        # Get product rating
        rating_div = container.find('div', class_="DisplayRating__RatingTag-sc-1mjg3vv-0 hOkpqp")
        rating = rating_div.text.strip() if rating_div else "N/A"

        # Get product price
        price_span = container.find('span', class_="price-label")
        price = price_span.text.strip() if price_span else "N/A"

        veg_nonveg_div = container.find('div', id="nonVegBoxWrapper")
        if veg_nonveg_div:
            veg_nonveg = "Non-Veg"
        else:
            veg_nonveg = "Veg"
        
        # Add the product details to the list
        products.append({
            "Name": name,
            "Description": description,
            "Rating": rating,
            "Price": price,
            "Veg/Non-Veg": veg_nonveg
        })

    # Print all products
    for product in products:
        print(f"Name: {product['Name']}")
        print(f"Description: {product['Description']}")
        print(f"Rating: {product['Rating']}")
        print(f"Price: {product['Price']}")
        print(f"Veg/Non-Veg: {product['Veg/Non-Veg']}")
        print("\n" + "="*40 + "\n")
        
shahi_birayni_df = pd.DataFrame(products)
print(shahi_birayni_df)
```

```python
import re
import pandas as pd

products = []
p25 = p24.find_all('figcaption')
p26 = p24.find_all('div', class_=re.compile(r"^style__CardBottom"))

for p1, p2 in zip(p25, p26):
    # Extract title
    title_div = p1.find('div', class_=re.compile(r"^style__Title"))
    title = title_div.text if title_div else "No Title"

    # Extract description
    desc_p = p1.find('p')
    desc = desc_p.text if desc_p else "No Description"

    # Extract rating
    rating_div = p1.find('div', class_=re.compile(r"^DisplayRating"))
    rating = rating_div.text if rating_div else "No Rating"

    # Extract price
    price_span = p2.find('span')
    price = price_span.text if price_span else "No Price"

    # Determine Veg/Non-Veg
    veg_div = p1.find('div', {'id': 'vegBoxWrapper'})
    veg_nonveg = "Veg" if veg_div else "Non-Veg"

    products.append({
        "Name": title,
        "Description": desc,
        "Rating": rating,
        "Price": price,
        "Veg/Non-Veg": veg_nonveg
    })

    # Debugging output
    print(f"Title: {title}")
    print(f"Description: {desc}")
    print(f"Rating: {rating}")
    print(f"Price: {price}")
    print(f"Veg/Non-Veg: {veg_nonveg}")

# Convert the list of product details to a pandas DataFrame
daastan_e_kebabs_df = pd.DataFrame(products)
```

```python
p27 = root4.find('div', class_=re.compile(r"^stickyHeader"))
p28 = p27.find('div', class_=re.compile(r"^style__Wrapper"))
p29 = p28.find('div', class_=re.compile(r"^style__Container"))
p30 = p29.find('div', class_=re.compile(r"^sc-hKwCoD"))

p31 = p30.find_all('figcaption')
p32 = p30.find_all('div', class_=re.compile(r"^style__CardBottom"))
```

```python
import pandas as pd
import re

# Assuming p31 and p32 contain the relevant data
products = []
for p1, p2 in zip(p31, p32):
    # Extract title
    title_div = p1.find('div', class_=re.compile(r"^style__Title"))
    title = title_div.text if title_div else "No Title"

    # Extract description
    desc_p = p1.find('p')
    desc = desc_p.text if desc_p else "No Description"

    # Extract rating
    rating_div = p1.find('div', class_=re.compile(r"^DisplayRating"))
    rating = rating_div.text if rating_div else "No Rating"

    # Extract price
    price_span = p2.find('span')
    price = price_span.text if price_span else "No Price"

    # Determine Veg/Non-Veg
    veg_nonveg_svg = p1.find('div', class_=re.compile(r"^style__Box")).find('svg')
    if veg_nonveg_svg:
        veg_nonveg_color = veg_nonveg_svg.find('path').get('stroke')
        veg_nonveg = "Veg" if veg_nonveg_color == "#008000" else "Non-Veg"
    else:
        veg_nonveg = "Unknown"

    # Append product details
    products.append({
        "Name": title,
        "Description": desc,
        "Rating": rating,
        "Price": price,
        "Veg/Non-Veg": veg_nonveg
    })

    # Debugging output
    print(f"Title: {title}")
    print(f"Description: {desc}")
    print(f"Rating: {rating}")
    print(f"Price: {price}")
    print(f"Veg/Non-Veg: {veg_nonveg}\n\n")

# Convert the list of product details to a pandas DataFrame
newly_launched_df = pd.DataFrame(products)
```

```python
# Assuming `root6` contains the parsed HTML content of the page
p33 = root6.find('div', class_=re.compile(r"^stickyHeader"))
p34 = p33.find('div', class_=re.compile(r"^style__Wrapper"))
p35 = p34.find('div', class_=re.compile(r"^style__Container"))
p36 = p35.find('div', class_=re.compile(r"^sc-hKwCoD"))
```

```python
p37 = p36.find_all('figcaption')
p38 = p36.find_all('div', class_=re.compile(r"^style__CardBottom"))

products = []
for p1, p2 in zip(p37, p38):
    # Extract title
    title_div = p1.find('div', class_=re.compile(r"^style__Title"))
    title = title_div.text.strip() if title_div else "No Title"

    # Extract description
    desc_p = p1.find('p', class_=re.compile(r"^style__Desc"))
    desc = desc_p.text.strip() if desc_p else "No Description"

    # Extract rating
    rating_div = p1.find('div', class_=re.compile(r"^DisplayRating"))
    rating = rating_div.text.strip() if rating_div else "No Rating"

    # Extract price
    price_span = p2.find('span', class_=re.compile(r"^price-label"))
    price = price_span.text.strip() if price_span else "No Price"

    # Extract Veg/Non-Veg
    veg_div = p1.find('div', id="vegBoxWrapper")
    if veg_div:
        veg_svg = veg_div.find('svg')
        if veg_svg and veg_svg.find('circle', fill="#008000"):
            veg_nonveg = "Veg"
        else:
            veg_nonveg = "Non-Veg"
    else:
        veg_nonveg = "No Info"

    # Append product details to the list
    products.append({
        "Name": title,
        "Description": desc,
        "Rating": rating,
        "Price": price,
        "Veg/Non-Veg": veg_nonveg,
    })

    # Debugging output
    print(f"Title: {title}")
    print(f"Description: {desc}")
    print(f"Rating: {rating}")
    print(f"Price: {price}")
    print(f"Veg/Non-Veg: {veg_nonveg}\n")

# Convert the list of product details to a pandas DataFrame
desserts_and_beverages_by_behrouz_df = pd.DataFrame(products)
```
